# Data Preparation



In order to fit the categorical attributes into the correlation matrix and test for statistical significance, I had to convert some key observations into binary values (i.e 1 and 0):

NAME_CONTRACT_TYPE = "Cash loans"--> 1, otherwise 0

CODE_GENDER = "M" --> 1, otherwise 0

FLAG_OWN_CAR = "N" --> 1, otherwise 0

FLAG_OWN_REALTY = "N" --> 1, otherwise 0

NAME_TYPE_SUITE = "Unaccompanied" --> 1, otherwise 0

NAME_INCOME_TYPE = "Working" --> 1, otherwise 0

NAME_EDUCATION_TYPE = "Secondary / secondary special" --> 1, otherwise 0

NAME_HOUSING_TYPE = "House / apartment" --> 1, otherwise 0

PREV_CONTRACT_STATUS = "Approved" --> 0, otherwise 1

Most of the other features unfortunately have too much cardinality to easily convert into binary values

```python
Cred_Default['NAME_CONTRACT_TYPE'] = (Cred_Default['NAME_CONTRACT_TYPE'] == "Cash loans").astype("int64")
Cred_Default['CODE_GENDER'] = (Cred_Default['CODE_GENDER'] == "M").astype("int64")
Cred_Default['FLAG_OWN_CAR'] = (Cred_Default['FLAG_OWN_CAR'] == "N").astype("int64")
Cred_Default['FLAG_OWN_REALTY'] = (Cred_Default['FLAG_OWN_REALTY'] == "N").astype("int64")
Cred_Default['NAME_TYPE_SUITE'] = (Cred_Default['NAME_TYPE_SUITE'] == "Unaccompanied").astype("int64")
Cred_Default['NAME_HOUSING_TYPE'] = (Cred_Default['NAME_HOUSING_TYPE'] == "House / apartment").astype("int64")
Cred_Default['PREV_CONTRACT_STATUS'] = (Cred_Default['PREV_CONTRACT_STATUS'] != "Approved").astype("int64")
Cred_Default['NAME_INCOME_TYPE'] = (Cred_Default['NAME_INCOME_TYPE'] == "Working").astype("int64")
Cred_Default['NAME_EDUCATION_TYPE'] = (Cred_Default['NAME_EDUCATION_TYPE'] == "Secondary / secondary special").astype("int64")

Cred_Default.head()
```

# Model Building

```python
# These 15 features have been selected based on the EDA_FeatureEngineering Python code
model_1 = Cred_Default[['TARGET', 'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'AMT_CREDIT',  'DAYS_BIRTH', 'FLAG_EMP_PHONE', 'DAYS_ID_PUBLISH','REGION_RATING_CLIENT_W_CITY', 'LIVE_CITY_NOT_WORK_CITY', 'REG_CITY_NOT_LIVE_CITY', 'DAYS_LAST_PHONE_CHANGE', 'PREV_CONTRACT_STATUS','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE', 'DOCUMENTS_GIVEN']] 
```

```python
model_1.isnull().sum()
```

```python
model_1.tail()
```

```python
X, y = model_1.iloc[:, model_1.columns != 'TARGET'], model_1.iloc[:, 0] # X - include all features until the end, Y - include only 1st column which is TARGET
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state = 42)
```

Stratified Train-Test split so the set distributions are the same as the population distribution.

```python
# Normalise feature values in train data using standard scaler

scaler = StandardScaler()
normalised_x_train = scaler.fit_transform(X_train)
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train == 0)))
  
# import SMOTE module from imblearn library and generate synthetic data points
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1.0, random_state = 42)
X_train_res, y_train_res = sm.fit_resample(normalised_x_train, y_train)
  
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
  
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))
```

```python
print(X_train_res)
```

```python
X_test = scaler.transform(X_test)
```

Using SMOTE, we have now oversampled the minority class to be balanced with the majority class. We have also normalised the data using the train dataset statistics, so that all values operate on the same scale. None of the information from the test data has leaked into the training.

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, auc, accuracy_score
```

```python
logreg = LogisticRegression(C=1.0, penalty='l2', solver="lbfgs", max_iter=100)
logreg.fit(X_train_res, y_train_res)
# Need to consider hyperparameters as well such as:
# HP1 - C - The inverse of regularization strength (smaller number prevents overfitting)
# HP2 - Solver - algorithm to use for optimization
# HP3 - Max Iterations
```

```python
# probs = logreg.predict_proba(X_test)[:, 1]
# threshold = 0.445

# y_pred = np.where(probs >= threshold, 1, 0) - this chunk of code allows for sensitivity testing as the probability thresholds can be changed

# Sensitivity Test (0.445, 50% Accuracy & 74% Recall -  0.36, 30% Accuracy & 89% Recall, 0.185, 10% Accuracy & 100% Recall)
```

```python
y_pred = logreg.predict(X_test)
logreg.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm_lr = confusion_matrix(y_test, y_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=logreg.classes_)

disp.plot()

plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Calculate the quintiles of the probabilities for overall dataset
Overall_probs = logreg.predict_proba(X_test)
Overall_quintiles = np.percentile(Overall_probs, [20, 40, 60, 80, 100])

print("20th Percentile:", Overall_quintiles[0])
print("40th Percentile:", Overall_quintiles[1])
print("60th Percentile:", Overall_quintiles[2])
print("80th Percentile:", Overall_quintiles[3])
print("100th Percentile:", Overall_quintiles[4])
```

Borrower risk profiling/segmentation that goes beyond default & non-default classifications. See AT3 report for more details

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score

# Define the hyperparameter grid to search
param_grid = {'C': [0.1, 1, 10, 100],
              'penalty': ['l1', 'l2'],
              'solver': ['newton-cg', 'lbfgs', 'liblinear']}

# Create a logistic regression estimator
logreg = LogisticRegression()

# Define a scoring metric
scoring = {'Accuracy': make_scorer(accuracy_score)}

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=logreg, param_grid=param_grid, scoring="balanced_accuracy", cv=5)

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train_res, y_train_res)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best hyperparameters:", best_params)

# Get the best score
best_score = grid_search.best_score_
print("Best accuracy: {:.2f}%".format(best_score * 100))

# Get the best estimator
best_estimator = grid_search.best_estimator_

# Use the best estimator to make predictions on the test data
y_pred = best_estimator.predict(X_test)
```

It was found that search for the hyperparameters in logistic regression was far too computationally and time intensive. However, based on our research of the different solvers and trial & error, we believe the above HPs are optimal.

```python
# Define the probabilistic range
lower_bound = 0.46
upper_bound = 0.5

# Create a boolean formula to determine which observations fall within the specified range
in_range = np.logical_and(Overall_probs >= lower_bound, Overall_probs < upper_bound)

# Count the number of observations that fall within the range. Divide by 2 as predict_proba prints probability for both classes - 0 to 1.0 is 61503 observations.
num_in_range = np.sum(in_range) / 2

# Print the result
print(f"Number of observations in range: {num_in_range}")
```

0.46 - 0.5 was the probability range used in the AT3 presentation (just below the threshold). Credit team have to be mindful of how many observations fall within this range, and determine whether this fits within their risk appetite for operational efficiency.
